# Feature Insertion

```python
import pandas as pd 

tourist_destinations_baguio_df = pd.read_csv(r'../csv/uncleaned/tourist_destinations_baguio.csv')
tourist_destinations_baguio_df = tourist_destinations_baguio_df.sort_values(['total_reviews'], ascending=[0]).reset_index(drop=True)
tourist_destinations_baguio_df
```

```python
t10_place_list = list(tourist_destinations_baguio_df["place"][:10])
t10_place_list[6] = 'Atonement Cathedral'
t10_place_list[8] = 'BenCab Museum'
t10_place_list
```

```python
import pandas as pd

data = pd.read_csv('../csv/cleaned/cleaned_variations_review_posts2.csv')
data['lemmatized_words_review_post'].replace('', pd.np.nan, inplace=True)
data.dropna(subset=['lemmatized_words_review_post'], inplace=True)
data
```

```python
top10_destinations_reviews_df = pd.DataFrame()

for destination in t10_place_list:
    destinations_reviews_df = data[data['tourist_destination'].str.contains(str(destination))]
    top10_destinations_reviews_df = top10_destinations_reviews_df.append(destinations_reviews_df, ignore_index=True)

top10_destinations_reviews_df
```

```python
data2 = top10_destinations_reviews_df.to_dict()
```

# Insert Relevant Pos Tags

```python
import spacy

selected_tags = ["NOUN", "ADJ", "ADV", "PROPN", "VERB", "X"]
tag_compil = {"NOUN" : [], "ADJ" : [], "ADV" : [], "PROPN" : [], "VERB" : [], "X" : []}

for idx, value in enumerate(list(data2['lemmatized_words_review_post'].values())):
    print("Loop: ", idx)
    sp = spacy.load('en_core_web_sm')
    sen = sp(value)

    pos_tagged = []
    selected_tags = {"NOUN" : 0, "ADJ" : 0, "ADV" : 0, "PROPN" : 0, "VERB" : 0, "X" : 0}

    for word in sen:
        if word.pos_ in selected_tags.keys():
            selected_tags[str(word.pos_)] += 1
            #pos_tagged.append((word.text, word.pos_))
    
    for idx, value in tag_compil.items():
        tag_compil[str(idx)].append(selected_tags[str(idx)])

tag_compil
```

```python
new = pd.DataFrame.from_dict(tag_compil)

new.to_csv("../csv/uncleaned/pos_tags.csv", index = False)
new
```

```python
for tag in selected_tags:
    top10_destinations_reviews_df[str(tag)] = tag_compil[str(tag)]

top10_destinations_reviews_df
```

# Sentiment Analysis

```python
from textblob import TextBlob

def analyze_score(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'

top10_destinations_reviews_df['subjectivity'] = top10_destinations_reviews_df['lemmatized_words_review_post'].map(lambda x: TextBlob(x).sentiment.subjectivity)
top10_destinations_reviews_df['polarity'] = top10_destinations_reviews_df['lemmatized_words_review_post'].map(lambda x: TextBlob(x).sentiment.polarity)
top10_destinations_reviews_df['sent_analysis'] = top10_destinations_reviews_df['polarity'].map(lambda x: analyze_score(x))

top10_destinations_reviews_df.drop(['subjectivity', 'polarity'], axis=1, inplace = True)
top10_destinations_reviews_df
```

```python
top10_destinations_reviews_df.to_csv("../csv/cleaned/cleaned_variations_review_posts_features.csv", index = False)
top10_destinations_reviews_df
```

# Sentiment Analysis Summary

```python
top10_destinations_reviews_df = pd.read_csv(r'../csv/cleaned/cleaned_variations_review_posts_features.csv')
```

```python
sentiments = top10_destinations_reviews_df['sent_analysis'].unique()
tourist_destinations = top10_destinations_reviews_df['tourist_destination'].unique()
tourist_destinations[6] = 'Atonement Cathedral'
```

```python
sentiments = {}

for tourist_destination in tourist_destinations:
    contain_values = top10_destinations_reviews_df[top10_destinations_reviews_df['tourist_destination'].str.contains(tourist_destination)]
    #display(pd.DataFrame(contain_values.sent_analysis.value_counts()))
    sentiments[str(tourist_destination)] = []
    for idx, name in enumerate(contain_values.sent_analysis.value_counts().tolist()):
        sentiments[str(tourist_destination)].append(contain_values.sent_analysis.value_counts()[idx])
#sentiments

positive = []
neutral = []
negative = []

for idx, value in sentiments.items():
    positive.append(value[0])
    neutral.append(value[1])
    negative.append(value[2])

```

```python
sentiment_analysis_df = pd.DataFrame(zip(tourist_destinations, positive, neutral, negative), columns = ['Tourist Destination', 'Positive', 'Neutral', 'Negative'])
sentiment_analysis_df
```
