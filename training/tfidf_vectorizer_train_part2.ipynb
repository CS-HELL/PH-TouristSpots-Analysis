```python
import pandas as pd 

tourist_destinations_baguio_df = pd.read_csv(r'../csv/uncleaned/tourist_destinations_baguio.csv')
tourist_destinations_baguio_df = tourist_destinations_baguio_df.sort_values(['total_reviews'], ascending=[0]).reset_index(drop=True)
tourist_destinations_baguio_df
```

```python
t10_place_list = list(tourist_destinations_baguio_df["place"][:10])
t10_place_list[6] = 'Atonement Cathedral'
t10_place_list[8] = 'BenCab Museum'
t10_place_list
```

```python
import pandas as pd

data = pd.read_csv('../csv/cleaned/cleaned_variations_review_posts_features.csv')
data
```

```python
data.tourist_destination.value_counts()
```

```python
top10_destinations_reviews_df = pd.DataFrame()

for destination in t10_place_list:
    destinations_reviews_df = data[data['tourist_destination'].str.contains(str(destination))]
    top10_destinations_reviews_df = top10_destinations_reviews_df.append(destinations_reviews_df, ignore_index=True)

top10_destinations_reviews_df
```

```python
top10_destinations_reviews_df.sent_analysis = pd.Categorical(top10_destinations_reviews_df.sent_analysis)
top10_destinations_reviews_df['sent_analysis_cat'] = top10_destinations_reviews_df.sent_analysis.cat.codes

top10_destinations_reviews_df.drop(['sent_analysis'], axis=1, inplace = True)
top10_destinations_reviews_df
```

```python
data = top10_destinations_reviews_df
tourist_destinations = data['tourist_destination']
print(len(tourist_destinations.unique()))
```

```python
tourist_destinations.unique()
```

# Train

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy

tfv = TfidfVectorizer(stop_words = 'english')
tfv_text = tfv.fit_transform(data['lemmatized_words_review_post'].values.astype('U'))

# For Machine Based
from sklearn.model_selection import train_test_split
vectors_train, vectors_test, dest_train, dest_test = train_test_split(tfv_text, tourist_destinations, test_size = 0.1, random_state = 50)
```

```python
x = tfv_text.toarray()

# For the template for input
df_tfidf_model = pd.DataFrame(x, columns = tfv.get_feature_names())
df_tfidf_model['tourist_destination'] = tourist_destinations
df_tfidf_model
```

# Support Vector Machine

```python
from sklearn import svm
svm = svm.SVC(kernel='rbf', probability=True)
svm.fit(vectors_train, dest_train)
```

```python
dest_pred = svm.predict(vectors_test)
actual_predict = pd.DataFrame()
actual_predict["actual"] = dest_test
actual_predict["predict"] = dest_pred
actual_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(dest_test, dest_pred))

# calculate accuracy of class predictions
from sklearn import metrics
print(f'Accuracy -> ', metrics.accuracy_score(dest_test, dest_pred))
```

```python
import pickle

# save the classifier
with open('../models/svm_classifier.pkl', 'wb') as fid:
    pickle.dump(svm, fid)    
```

```python
import pickle

# load it again
with open('../models/svm_classifier.pkl', 'rb') as fid:
    svm = pickle.load(fid)
```
